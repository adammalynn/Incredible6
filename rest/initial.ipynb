```python
import matplotlib.pyplot as plt
import json
import pandas as pd
import numpy as np
import requests
import csv
from pprint import pprint
import traceback
import re
from config import api_key
```

```python
def search_resty(offset, city='raleigh', state='NC'):
   url = 'https://api.yelp.com/v3/businesses/search'
   headers = {
       'Authorization': 'Bearer {}'.format(api_key),
   }
   url_params = {
           "categories": 'Restaurants',
           "location": city,
           "state": state,
           'offset': offset,  # We are going to iterate the offset
           'limit': 50  # Maximum return of results per request (ref: API documentation).
            }

   response = requests.get(url, headers=headers, params=url_params)
   return response.json() # Returns a JSON.
```

```python
if __name__ == "__main__":
    cities = ['asheville','raleigh']
    
    with open('../restaurant5.csv', 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile, delimiter=',')
        csvwriter.writerow(['City', 'Name', 'Review', 'Zip_codes', 'Rating', 'Price', 'Lat', 'Long', 'Title'])
        for city in cities:
            restaurant_data = []
            for index in range(0,600,50):
                api_response = search_resty(index, city)
                for res in api_response['businesses']:
                    try:
                        restaurant_data.append([city, res['name'],
                                    res['review_count'],
                                    res['location']['zip_code'],
                                    res['rating'],
                                    res['price'],
                                    res['coordinates']['latitude'],
                                    res['coordinates']['longitude'],
                                    res['categories'][0]['title']])
        
                    except KeyError:
                        print(f"{res['name']} Restaurant data is missing")
            for res_value in restaurant_data:
                try:
                    list1 = []
                    for value in res_value:
                        if isinstance(value,str):
                            value = re.sub('\u200b', '', value)
                        list1.append(value)
                    csvwriter.writerow(list1)
                except Exception as e:
                    print(res_value)
                    traceback.print_exc(e)
```
